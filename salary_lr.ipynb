```python
#importing neccesary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
data = pd.read_csv('/content/Salary_Data.csv')
data
```

```python
#separation of dependent and independendant variables
#independant variable must be a 2d array or two dimensional array
x = data.iloc[:,:1]
x
```

```python
#dependent variable must be a series or one dimentional array
y = data["Salary"]
y
```

```python
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
#train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#model building
#train the model using trained data
#test the model on trained data
#evaluate the model
from  sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#test the model
y_pred = lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
#Evaluation of model
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,lr.predict(x_train),color="red")
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,lr.predict(x_test),color="red")
```
